```python
missing_values_count = 0
unexpected_values_count = 0

season_name = []

for s in data1["month_column"]:
    if pd.isna(s):
        season_name.append("Unknown")
        missing_values_count += 1
    elif s in ["January", "February"]:
        season_name.append("Winter")
    elif s in ["March", "April"]:
        season_name.append("Spring")
    elif s in ["May", "June"]:
        season_name.append("Summer")
    elif s in ["July", "August"]:
        season_name.append("Monsoon")
    elif s in ["September", "October"]:
        season_name.append("Autumn")
    elif s in ["November", "December"]:
        season_name.append("Pre-Winter")
    else:
        season_name.append("Unknown")
        unexpected_values_count += 1

data1["season"] = season_name

print(f"Missing values: {missing_values_count}")
print(f"Unexpected values: {unexpected_values_count}")
```
